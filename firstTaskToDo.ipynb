### Регистрируемся/Устанавливаем Git на локальный пк
```
    Создаем на Git репозиторий например firstTask.git

   **далее в cmd**
# Авторизуемся в Git на ПК (или через браузер)
git config --global user.name "ВашеИмя"
git config --global user.email "ваш@email.com"
# Инициализируем папку с файлом
cd C:\Users\Пользователь\Jupyter
git init
# Связываем удаленный репозиторий firstTask.git с именем origin
git remote add origin https://github.com/ВашеИмя/firstTask.git
git remote -v
# Отвязываем удаленный репозиторий firstTask.git с именем origin
git remote remove origin
git remote -v
# Проверка статуса:
git status
# Добавление файлов в коммит:
git add .
# Или 1 конкретный файл:
git add firstTask.ipynb
# Создаем новую ветку
git checkout -b main
# Получение изменений с GitHub:
git pull origin main
# Создание коммита:
git commit -m "Ваше сообщение о коммите"
# Отправка изменений на GitHub:
git push origin main

### После этого можно скопировать ссылку на файл из вашего гит для преподавателя
https://github.com/ВашеИмя/firstTask/blob/main/firstGitFile.ipynb


📋 Полная таблица команд Git с описаниями
🔧 Основные команды
Команда	Описание	Пример
git init	Инициализирует новый репозиторий	git init
git clone <url>	Клонирует удаленный репозиторий	git clone https://github.com/user/repo.git
git status	Показывает статус файлов	git status
git add <file>	Добавляет файлы в индекс	git add .
git commit -m "msg"	Создает коммит	git commit -m "Add feature"
git push	Отправляет коммиты на сервер	git push origin main
git pull	Забирает изменения с сервера	git pull origin main
📁 Работа с файлами
Команда	Описание	Пример
git add .	Добавляет все файлы	git add .
git add <file>	Добавляет конкретный файл	git add index.html
git rm <file>	Удаляет файл из индекса и рабочей директории	git rm oldfile.txt
git mv <old> <new>	Переименовывает или перемещает файл	git mv old.txt new.txt
git restore <file>	Отменяет изменения в файле	git restore script.py
git restore --staged <file>	Убирает файл из индекса	git restore --staged file.js
📊 Просмотр истории
Команда	Описание	Пример
git log	Показывает историю коммитов	git log
git log --oneline	Краткая история коммитов	git log --oneline
git log -p	История с изменениями в файлах	git log -p
git show <commit>	Показывает конкретный коммит	git show abc123
git diff	Показывает незакоммиченные изменения	git diff
git diff --staged	Показывает изменения в индексе	git diff --staged
🌿 Ветвление
Команда	Описание	Пример
git branch	Список веток	git branch
git branch <name>	Создает новую ветку	git branch feature
git checkout <branch>	Переключается на ветку	git checkout main
git checkout -b <branch>	Создает и переключается на ветку	git checkout -b feature
git merge <branch>	Объединяет ветки	git merge feature
git branch -d <branch>	Удаляет ветку	git branch -d feature
git switch <branch>	Переключение между ветками	git switch main
🔄 Отмена изменений
Команда	Описание	Пример
git reset <file>	Убирает файл из индекса	git reset file.txt
git reset --soft <commit>	Откатывает к коммиту, сохраняя изменения	git reset --soft HEAD~1
git reset --hard <commit>	Полный откат к коммиту	git reset --hard abc123
git revert <commit>	Создает новый коммит, отменяющий изменения	git revert abc123
git clean -fd	Удаляет неотслеживаемые файлы	git clean -fd
🛠 Удаленные репозитории
Команда	Описание	Пример
git remote -v	Показывает удаленные репозитории	git remote -v
git remote add <name> <url>	Добавляет удаленный репозиторий	git remote add origin https://...
git remote remove <name>	Удаляет удаленный репозиторий	git remote remove origin
git fetch <remote>	Загружает изменения без слияния	git fetch origin
git push -u <remote> <branch>	Отправляет ветку и устанавливает upstream	git push -u origin main
git push --delete <remote> <branch>	Удаляет ветку на удаленном репозитории	git push --delete origin old-branch
🏷 Теги
Команда	Описание	Пример
git tag	Показывает все теги	git tag
git tag <name>	Создает легковесный тег	git tag v1.0
git tag -a <name> -m "msg"	Создает аннотированный тег	git tag -a v1.0 -m "Version 1.0"
git push --tags	Отправляет теги на сервер	git push --tags
git tag -d <name>	Удаляет тег	git tag -d v1.0
🔍 Поиск
Команда	Описание	Пример
git grep "text"	Ищет текст в файлах	git grep "TODO"
git log --grep="text"	Ищет коммиты по сообщению	git log --grep="bug"
git log -S "text"	Ищет коммиты по изменению кода	git log -S "function_name"
⚙ Конфигурация
Команда	Описание	Пример
git config --list	Показывает настройки	git config --list
git config user.name	Показывает имя пользователя	git config user.name
git config --global user.name "name"	Устанавливает глобальное имя	git config --global user.name "John"
git config --global user.email "email"	Устанавливает глобальный email	git config --global user.email "john@example.com"
🗂 Справка
Команда	Описание	Пример
git help <command>	Показывает справку по команде	git help push
git <command> -h	Краткая справка по команде	git push -h
🎯 Полезные алиасы
Команда	Описание
git lg	Красивый лог (нужно настроить алиас)
git st	Сокращение для status
git co	Сокращение для checkout
git br	Сокращение для branch
💡 Самые частые команды:
bash
# Ежедневный workflow
git status                  # Что изменилось?
git add .                   # Добавить все файлы
git commit -m "Сообщение"   # Закоммитить изменения
git push origin main        # Отправить на сервер

# Получение обновлений
git pull origin main        # Получить изменения

# Работа с ветками
git checkout -b feature     # Создать новую ветку
git merge feature           # Влить ветку в основную
```

### DOCKER
```
После установки Docker Desktop на Windows 10, Docker попросит вас обновить/включить wsl. Если wsl не включен, то включаем:
    dism.exe /online /enable-feature /featurename:Microsoft-Windows-Subsystem-Linux /all /norestart    # Включение WSL через DISM из cmd (от имени администратора)
    Enable-WindowsOptionalFeature -Online -FeatureName Microsoft-Windows-Subsystem-Linux               # Включение WSL через PowerShell  (от имени администратора)
обязательно перезагрузите пк после чего можно поставить например Ubuntu на ваш wsl:
    wsl --install -d Ubuntu
Теперь можно добавить контейнер с jupyter в докер. Для этого делать вообще ничего особого не надо:
в cmd переходим в папку ipynb_files с файлом docker-compose.yml (из презентации) вводим команду в cmd
    docker compose up -d
ждем пока докер сам все скачает и собственно все
```


### Сравнение версии jupyter из контейнера докера и jupyter из анаконды
```
Чтобы сравнить версии jupyter из контейнера докера и jupyter из анаконды нужно поменять порт в файле docker-compose.yml
Для этого останавливаем контейнер, в cmd
    docker compose down

Редактируем файл docker-compose.yml меняем порт "8888:8888" на "8890:8888":
services:
  lab:
    image: jupyter/scipy-notebook:latest
    ports:
      - "8890:8888"
...

сохраняем, после чего заново запускаем контейнер в cmd
    docker compose up -d

```
