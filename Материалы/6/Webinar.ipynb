```python
import pandas as pd

# создаем датафрейм с колонкой Salary  со значением [114000, 50000, 100000]
# 1) преобразуем их в строки вида "45000 руб" и сохраним в новой колонке
# 2) из полученных строк извелечем обратно числа и приведем их к типу int в новой колонке


df = pd.DataFrame({"Salary": [114000, 50000, 100000]})  

df["Salary2"] = df["Salary"].astype(str) + " руб"
print(df["Salary2"],'\n')

df["Salary3"] = df["Salary2"].str.extract(r'(\d+)').astype(int)
#df["Salary3"] = df["Salary2"].str.extract(r'(\d+)').astype(int)

print(df["Salary3"])

df
```

```python
# DataFrame с колонкой Record со строками формата -  (3 строки)
#"id=105; user=Alice; score=92"
#"id=106; user=Bob; score=88"
#"id=107; user=Carol; score=95"
# извлечь user и score в отдельные колонки

# Исходный DataFrame
df = pd.DataFrame({
    "Record": [
        "id=105; user=Alice; score=92",
        "id=106; user=Bob; score=88", 
        "id=107; user=Carol; score=95"
    ]
})

split_df = df['Record'].str.split('; ', expand=True)
df['user'] = split_df[1].str.split('=').str[1]
df['score'] = split_df[2].str.split('=').str[1]

# Извлечение user и score с помощью регулярного выражения
df[['user', 'score']] = df['Record'].str.extract(r'user=([^;]+); score=(\d+)')
df[['user', 'score']] = df['Record'].str.extract(r""r'user=([^;]+); score=(\d+)')
df
```

```python
# DataFrme с колонкой Record со строкоами формата - "id=105; user=Alice; score=92",
# "id=106; user=Bob; score=88",
# "id=107; user=Carol; score=95"
# извлечь user и score в отдельные колонки

df = pd.DataFrame({
"Record": [
"id=105; user=Alice; score=92",
"id=106; user=Bob; score=88",
"id=107; user=Carol; score=95"
]
})
df


extracted = df["Record"].str.extract(r"user=(?P<user>[^;]+);\s*score=(?P<score>\d+)") # именованные группы в регулярке
extracted


extracted["score"] = extracted["score"].astype(int)

df = pd.concat([df, extracted], axis = 1)
df


#df.dtypes
```

```python
# преобразуем список словарей [{"City: "Moscow", "Population": 15},{},{}] в датафрейм
# сделаем City - индексом и отсортируем по Population

data = [
    {"City": "Moscow", "Population": 15},
    {"City": "Kazan", "Population": 1.5},
    {"City": "Spb", "Population": 5.5},
]

df = pd.DataFrame(data)
df = df.set_index('City')
df = df.sort_values('Population', ascending=False)

df
```
