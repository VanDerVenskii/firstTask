```python
import pandas as pd
import matplotlib.pyplot as plt

# Создадим простую таблицу с данными о продажах
data = {"Месяц": ["Янв", "Фев", "Мар", "Апр"], "Продажи": [10, 14, 12, 18]}
df = pd.DataFrame(data)
print("Наша таблица:\n", df)

# Посчитаем простую статистику по столбцу "Продажи"
print("\nСтатистика по продажам:\n", df["Продажи"].describe())

# Построим график продаж по месяцам
df.plot(x="Месяц", y="Продажи", kind="line", marker="o", title="Продажи по месяцам")
plt.tight_layout()
plt.show()
```

```python
# Создание Series из списка
s = pd.Series([10, 20, 30, 40], name="Продажи")
print("Series:\n", s)

# Индекс и значения
print("\nИндекс:", s.index)     # метки строк (по умолчанию от 0 до n-1)
print("Значения:", s.values)   # сами данные (массив NumPy)

# Обращение по индексу и по номеру
print("\nПо индексу 2:", s[2])         # значение на позиции 2
print("По срезу [1:3]:\n", s[1:3])    # подмножество значений

# Метаданные
print("\nИмя Series:", s.name)
print("Тип данных:", s.dtype)
```

```python
# Создание DataFrame из словаря
data = {
    "Категория": ["A", "A", "B", "B", "C"],
    "Продажи": [10, 15, 20, 25, 30],
    "Город": ["Москва", "СПб", "Москва", "СПб", "Казань"]
}
df = pd.DataFrame(data)
print("Таблица DataFrame:\n", df)

# Выборка столбца и строки
print("\nСтолбец 'Продажи':\n", df["Продажи"])
print("\nПервая строка (iloc[0]):\n", df.iloc[0])

# Группировка и агрегирование
grouped = df.groupby("Город")["Продажи"].mean()
print("\nСредние продажи по городам:\n", grouped)

# Описательные статистики по числовым данным
print("\nСтатистики:\n", df.describe())
```

```python
# Создадим таблицу
data = {
    "Категория": ["A", "A", "B", "B", "C", "C"],
    "Город": ["Москва", "СПб", "Москва", "СПб", "Москва", "СПб"],
    "Продажи": [10, 15, 20, 25, 30, 35]
}
df = pd.DataFrame(data)
print("Исходная таблица:\n", df)

# Сводная таблица: средние продажи по категориям и городам
pivot = pd.pivot_table(df, values="Продажи", index="Категория", columns="Город", aggfunc="mean")
print("\nСводная таблица:\n", pivot)

# Добавим ещё статистику: суммы продаж по городам
pivot_sum = pd.pivot_table(df, values="Продажи", index="Город", aggfunc="sum")
print("\nСуммарные продажи по городам:\n", pivot_sum)

# Быстрый просмотр корреляций между столбцами
print("\nКорреляции:\n", df.corr(numeric_only=True))
```

```python
import pandas as pd

# Расширим данные для наглядности
data = {
    "Категория": ["A", "A", "B", "B", "C", "C", "A", "B"],
    "Город": ["Москва", "СПб", "Москва", "СПб", "Москва", "СПб", "Москва", "Казань"],
    "Продажи": [10, 15, 20, 25, 30, 35, 12, 22],
    "Товар": ["Яблоки", "Груши", "Бананы", "Апельсины", "Виноград", "Персики", "Сливы", "Киви"]
}
df = pd.DataFrame(data)
print("Исходная таблица:\n", df)
```

```python
# 1. Объединение через запятую (аналог LISTAGG)
pivot_listagg = pd.pivot_table(
    df, 
    values="Товар", 
    index="Категория", 
    aggfunc=lambda x: ', '.join(x)
)
print("\nТовары по категориям (через запятую):\n", pivot_listagg)

# 2. В виде списка
pivot_list = pd.pivot_table(
    df, 
    values="Товар", 
    index="Категория", 
    aggfunc=list
)
print("\nТовары по категориям (списком):\n", pivot_list)

# 3. Сводная таблица с группировкой по двум измерениям
pivot_city_category = pd.pivot_table(
    df,
    values="Товар",
    index="Категория",
    columns="Город",
    aggfunc=lambda x: ', '.join(x),
    fill_value="нет товаров"
)
print("\nТовары по категориям и городам:\n", pivot_city_category)
```

```python
# Через groupby + agg
result_groupby = df.groupby('Категория')['Товар'].agg(lambda x: ', '.join(sorted(set(x)))).reset_index()
print("\nЧерез groupby:\n", result_groupby)

# Несколько агрегатных функций для строк
multi_agg = df.groupby('Категория').agg({
    'Товар': [', '.join, 'count', 'first', 'last', 'nunique'],
    'Город': [lambda x: ', '.join(sorted(set(x))), 'count', 'first', 'last', 'nunique']  # количество уникальных городов
})
print("\nНесколько агрегатных функций:\n", multi_agg)
```

```python
# Разные варианты агрегации строк
string_aggregations = df.groupby('Категория').agg({
    'Товар': [
        ', '.join,           # объединение через запятую
        ' | '.join,          # объединение через разделитель
        'count',             # количество записей
        'nunique',           # количество уникальных значений
        'first',             # первое значение
        'last',              # последнее значение
        'min',               # минимальное (по алфавиту)
        'max'                # максимальное (по алфавиту)
    ]
})
print("\nРазные агрегации для строк:\n", string_aggregations)
```

```python
# Создадим пример с настоящей корреляцией
import numpy as np

#np.random.seed(42)
data = {
    'Продажи': np.random.randint(100, 500, 50),
    'Реклама': np.random.randint(1000, 5000, 50),
    'Отзывы': np.random.normal(4.0, 0.5, 50)
}

#print(data)

df_corr = pd.DataFrame(data)

# Добавим зависимость между продажами и рекламой
df_corr['Продажи'] = df_corr['Реклама'] * 0.1 + np.random.normal(0, 50, 50)

print("Матрица корреляций:\n", df_corr.corr())
print(f"\nКорреляция Продажи-Реклама: {df_corr['Продажи'].corr(df_corr['Реклама']):.3f}")
```

```python
import pandas as pd

# Пример данных
data = {
    "Категория": ["A", "A", "A", "B", "B", "B", "C", "C"],
    "Город": ["Москва", "Москва", "СПб", "Москва", "СПб", "СПб", "Москва", "Казань"],
    "Товар": ["Яблоки", "Яблоки", "Груши", "Бананы", "Апельсины", "Бананы", "Виноград", "Персики"]
}
df = pd.DataFrame(data)

# Правильное использование нескольких агрегатных функций
multi_agg = df.groupby('Категория').agg(
    товар_unique=('Товар', lambda x: ', '.join(sorted(set(x)))),
    товар_count=('Товар', 'count'),
    товар_first=('Товар', 'first'),
    товар_last=('Товар', 'last'),
    товар_nunique=('Товар', 'nunique'),
    город_unique=('Город', lambda x: ', '.join(sorted(set(x)))),
    город_count=('Город', 'count'),
    город_first=('Город', 'first'),
    город_last=('Город', 'last'),
    город_nunique=('Город', 'nunique')
).reset_index()

print("Несколько агрегатных функций с именами:\n", multi_agg)
```

```python
# Способ 1: Через кортежи для многоуровневых колонок
multi_agg_tuples = df.groupby('Категория').agg({
    'Товар': [
        ('unique', lambda x: ', '.join(sorted(set(x)))),
        ('count', 'count'),
        ('first', 'first'),
        ('last', 'last'),
        ('nunique', 'nunique')
    ],
    'Город': [
        ('unique', lambda x: ', '.join(sorted(set(x)))),
        ('count', 'count'), 
        ('first', 'first'),
        ('last', 'last'),
        ('nunique', 'nunique')
    ]
})

print("Многоуровневые колонки:\n", multi_agg_tuples)
```

```python
# Определяем функции заранее
def unique_sorted(series):
    return ', '.join(sorted(set(series)))

# Используем именованные агрегации
result_clean = df.groupby('Категория').agg(
    Уникальные_товары=('Товар', unique_sorted),
    Всего_записей_товаров=('Товар', 'count'),
    Первый_товар=('Товар', 'first'),
    Последний_товар=('Товар', 'last'),
    Уникальных_товаров=('Товар', 'nunique'),
    Уникальные_города=('Город', unique_sorted),
    Всего_записей_городов=('Город', 'count'),
    Первый_город=('Город', 'first'),
    Последний_город=('Город', 'last'),
    Уникальных_городов=('Город', 'nunique')
).reset_index()

print("Чистый результат с понятными именами:\n", result_clean)
```
