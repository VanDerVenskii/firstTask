```python
import matplotlib.pyplot as plt

x = [i for i in range (-10, 11)]
y = [i**2 for i in x]

plt.plot(x, y, label="y = x ^ 2")

plt.title("hello matplotlib")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show()
```
