```python
import matplotlib.pyplot as plt
```

```python
# данные
x = [i for i in range(6)]
y1 = [i for i in x]
y2 = [i**2 for i in x]

fig, ax = plt.subplots()

ax.plot(x, y1, label="y = x")
ax.plot(x, y2, label="y = x**2")

ax.set_title("main is Figure/Axes")
ax.set_xlabel("ось X")
ax.set_ylabel("ось y")
ax.grid(True)
ax.annotate("выделим точку", xy=(2, 4), xytext=(3, 5),
            arrowprops=dict(arrowstyle="->"))

plt.show()
```

```python
# данные
x = [i for i in range(6)]
y1 = [i for i in x]
y2 = [i**2 for i in x]

# линейные диаграммы с разными стилями
plt.plot(x, y1, color="blue", marker="o", label="y = x")
plt.plot(x, y2, color="red", linestyle="--", marker="s", label="y = x**2")

plt.title("Линейная диаграмма")
plt.xlabel("Ось X")
plt.ylabel("Ось y")
plt.grid(True)

plt.show()
```

```python
np.random.seed(0) # фиксация генерации случайных чисел
x = np.random.rand(30) * 10 # 30 случайных точек от 0 до 10
y = 2 * x + np.random.rand(30) * 2 # линейная зависимость с шумом

# точечная диаграмма (scatter)
plt.scatter(x, y, color="purple", alpha=0.7, label="точки")

plt.title("Точечная диаграмма")
plt.xlabel("Ось X")
plt.ylabel("Ось Y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
categories = ["A", "B", "C", "D", "E"]
values = np.array([5, 7, 3, 8, 6]) # значения категорий

# Стобчатая диаграмма (Bar)
plt.bar(categories, values, color="skyblue", edgecolor="black", label="данные")

plt.title("Столбчатая диаграмма")
plt.xlabel("Категории")
plt.ylabel("Значения")
plt.legend()
plt.grid(axis="y")

plt.show()
```

```python
np.random.seed(0)
data = np.random.rand(500) # 500 случайных значений

# гистограмма
plt.hist(data, bins=15, color="orange", edgecolor="black", alpha=0.7, label="распределение")

plt.title("Гистограмма распределения")
plt.xlabel("Значения")
plt.ylabel("Частота")
plt.legend()
plt.grid(axis="y")

plt.show()
```

```python
np.random.seed(1)
data1 = np.random.normal(0, 1, 50)
data2 = np.random.normal(5, 2, 50)
data3 = np.random.normal(-2, 1.5, 50)

# Коробчатая диаграмма (boxplot)
plt.boxplot([data1, data2, data3], tick_labels=["Группа 1", "Группа 2", "Группа 3"])

plt.title("Коробчатая диаграмма")
plt.ylabel("Значения")
plt.grid(axis="y")

plt.show
```

```python
categories = ["A", "B", "C", "D"]
values = [40, 25, 20, 15]

# Секторная диаграмма
plt.pie(values, labels=categories, autopct="%.1f", startangle=90)

plt.title("Секторная диаграмма")
plt.show()
```

```python
categories = ["A", "B", "C", "D"]
values = [40, 25, 20, 15]

# настройка смещения сегментов
explode = [0.1, 0, 0, 0] # выделели первый сектор

plt.pie(values, labels=categories, autopct="%.1f%%",
        startangle=90, explode=explode, shadow=True)

plt.title("Секторная диаграмма с выделением сегмента")
```

```python
products = ("Ноутбук", "Смартфон", "Наушники", "монитор",
           "клавиатура", "Мышь", "Планшет", "Планшет") # товары (Ось Y)

data = [
    [15.0, 22.0, 10.0, 18.0, 9.0, 12.0, 8.0, 14.0], # онлайн
    [12.0, 8.0, 11.0, 15.0, 10.0, 14.0, 9.0, 7.0],  # Розница
    [7.0, 5.0, 9.5, 6.0, 12.5, 6.5, 10.0, 8.0],     # Партнеры
    [5.5, 6.5, 7.0, 9.0, 8.0, 7.0, 6.0, 5.5]        # Опт  
]

channels = 4 # число сегментов (каналы сбыта)
data = np.array(data)


fig, ax = plt.subplots(figsize=(10, 5))

bar_height = 0.5
opacity = 0.85
ind = np.arange(len(products))

legend_labels = ["Онлайн","Розница", "Партнеры", "Опт"]

for i in range(channels):
    if i == 0:
        rects = plt.barh(ind, data[i], height=bar_height, alpha=opacity, label=legend_labels[i])
    else:
        rects = plt.barh(
            ind,
            data[i],
            height=bar_height,
            alpha=opacity,
            left=data[:i].sum(axis=0),
            label=legend_labels[i]
        )
    for rect in rects:
        width = rect.get_width()
        cx = rect.get_x() + width / 2.0      
        cy = rect.get_y() + rect.get_height() / 2.0
        
        ax.annotate(
            f"{width:.1f}",
            xy=(cx, cy),
            xytext=(0, -10),
            textcoords="offset points",
            ha="center", va="bottom",
            fontsize=8, color="black"
        )



ax.set_title("Продажи по товарам и каналам")
ax.set_xlabel("Объем продаж")
ax.set_ylabel("Товары")
ax.grid(axis="x", linestyle=":", alpha=0.5)

plt.yticks(ind, products)

plt.legend(loc="upper right")

plt.tight_layout()

'''
# Все в одном (если rects_list уже определен)
[ax.annotate(f"{rect.get_width():.1f}", xy=(rect.get_x() + rect.get_width()/2, rect.get_y() + rect.get_height()/2),
             xytext=(0, -10), textcoords="offset points", ha="center", va="bottom", fontsize=8, color="black")
             for channel in rects_list for rect in channel]
             
Разберем параметры:
plt.barh(y, width, height=0.8, left=None, *, align='center', **kwargs)
plt.barh(ind, data[i], height=bar_height, alpha=opacity, left=cumulative_data[i], label=legend_labels[i])

y (в нашем случае ind): координаты по вертикальной оси, где будут расположены столбцы. Обычно это массив индексов (например, [0, 1, 2, ...] для каждого бара).
width (в нашем случае data[i]): ширина каждого бара (горизонтальный размер). Это массив значений, которые определяют, как далеко каждый бар простирается по горизонтали.
height: высота каждого бара (вертикальный размер). Может быть скаляром (тогда все бары одной высоты) или массивом. У нас bar_height=0.5.
left: горизонтальная координата начала каждого бара. По умолчанию 0. Мы используем cumulative_data[i] для stacked бара, чтобы каждый следующий сегмент начинался там, где закончился предыдущий.
alpha: прозрачность (от 0.0 до 1.0). У нас opacity=0.85.
label: метка для легенды. У нас берется из legend_labels[i].

Кроме этих, есть и другие параметры, которые можно передать в **kwargs:
color: цвет бара. Может быть строкой (например, 'red'), hex строкой (например, '#FF0000') или кортежем RGB.
edgecolor: цвет границы.
linewidth: ширина границы.
linestyle: стиль линии границы.
tick_label: метки для каждого бара (по умолчанию None, тогда используются значения из y).
xerr, yerr: ошибки (для горизонтальных бар ошибки по x это горизонтальные ошибки, но в barh они называются xerr, хотя по сути это ошибки по ширине).
capsize: размер колпачков на ошибках.
log: логарифмическая шкала (True/False).
orientation: ориентация, но для barh всегда горизонтальная.
Также можно передавать параметры из matplotlib.patches.Patch, например hatch (заливка), fill (заполнять или нет) и др.
В нашем примере мы используем необходимые для stacked bar chart параметры: y, width, height, left, alpha и label.
'''

plt.show()
```

```python
products = ("Ноутбук", "Смартфон", "Наушники", "монитор", "клавиатура", "Мышь", "Планшет", "Планшет") # товары (Ось Y)
data = (np.random.randint(11, 26, size=(4,8)) / 2.0)
fig, ax = plt.subplots(figsize=(10, 5))
ind = np.arange(len(products)) #cumulative_data = np.vstack([np.zeros_like(data[0]), np.cumsum(data, axis=0)])[:len(data)]
[plt.barh(ind, data[i], height=0.5, alpha=0.85, left=0 if i==0 else np.cumsum(data, axis=0)[i-1], label=["Онлайн","Розница", "Партнеры", "Опт"][i]) for i in range(len(data))]
[ax.bar_label(container, fmt='%.1f', label_type='center', fontsize=8, padding=3) for container in ax.containers]
ax.set(title="Продажи по товарам и каналам", xlabel="Объем продаж", ylabel="Товары", yticks=ind, yticklabels=products)
ax.grid(axis="x", linestyle=":", alpha=0.5)
ax.legend(loc="upper right")
fig.tight_layout()  # Автоматически подгоняет расположение элементов на графике
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Данные
products = ("Ноутбук", "Смартфон", "Наушники", "монитор", "клавиатура", "Мышь", "Планшет", "Планшет") # товары (Ось Y)
def generate_channel_data(size=8): return (np.random.randint(11, 26, size=size) / 2.0).tolist()
data = {
    'Онлайн': generate_channel_data(),
    'Розница': generate_channel_data(),
    'Партнеры': generate_channel_data(),
    'Опт': generate_channel_data()
}
fig, ax = plt.subplots(figsize=(10, 5))
pd.DataFrame(data, index=products).plot.barh(stacked=True, ax=ax, alpha=0.85)
[ax.bar_label(container, fmt='%.1f', label_type='center', fontsize=8) for container in ax.containers]
ax.set(title="Продажи по товарам и каналам", xlabel="Объем продаж", ylabel="Товары")
ax.grid(axis="x", linestyle=":", alpha=0.5)
plt.tight_layout()
plt.show()
```
