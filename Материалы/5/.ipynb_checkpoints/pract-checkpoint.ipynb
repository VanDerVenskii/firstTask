```python
age_counts = df['Age'].value_counts()
```

```python
# bar (столбчатая диаграмма) построим визуализацию
import matplotlib.pyplot as plt

plt.figure(figsize=(9,6))
plt.bar(age_counts.index, age_counts.values, color='red')

plt.show
```

```python
df = pd.read_csv('MoviesOnStreamingPlatforms.csv')
```

```python
print("Первые 10 строк:")
print(df.head(10))
```

```python
#Unnamed: 0","ID","Type

df.drop(columns=["Unnamed: 0","ID","Type"],inplace=True)

print(df.info())
```

```python
df.head(5)
```

```python
df["Age"]
```

```python
#заполним пропуски значением Unknown
# fillna
#df["Age"] = df["Age"].fillna("Unknown")
df["Age"].fillna("Unknown", inplace=True)
print(df.info())
```

```python
df.head(50)
```

```python
df["Rotten Tomatoes"] = pd.to_numeric(df["Rotten Tomatoes"].str.replace("/100", "")), errors='coerse'
print(df.info())
```

```python
df.head(5)
```

```python


platform_counts = {
    'Netflix': df["Netflix"].sum(),
    'Hulu': df["Hulu"].sum(),
    'Prime Video': df["Prime Video"].sum(),
    'Disney+': df["Disney+"].sum()
}

plt.figure(figsize=(9,6))
plt.bar(platform_counts.keys(), platform_counts.values(), color=['red', 'blue', 'green', 'gold'])
plt.title('Количество фильмов на каждой платформе')
plt.ylabel('Количество фильмов')
plt.xlabel('Платформа')
plt.show()
```

```python
# распределение фильмов по возрастным рейтингам Age в формате (bar)
# age_counts ... value_counts() ... визуализацию

# Получаем частоты значений в колонке 'Age'
age_counts = df['Age'].value_counts()

# Строим bar chart
plt.figure(figsize=(12, 6))
age_counts.plot(kind='bar', color=['red', 'blue', 'green', 'gold', 'purple', 'silver'], edgecolor='black')
plt.title('Распределение по возрастам')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.xticks(rotation=45) #, ha="left"
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()


age_counts_sorted = age_counts.sort_index()
age_counts_sorted.plot(kind='bar', color=['red', 'blue', 'green', 'gold', 'purple', 'silver'], edgecolor='black')
plt.title('Распределение по возрастам')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.xticks(rotation=45)
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15, 10))
fig.suptitle("Дашборд", fontsize=16)  # Исправлено: suptitle вместо subtitle

# Первый график - платформы
platform_counts = {
    'Netflix': df["Netflix"].sum(),
    'Hulu': df["Hulu"].sum(),
    'Prime Video': df["Prime Video"].sum(),
    'Disney+': df["Disney+"].sum()
}

ax[0].bar(platform_counts.keys(), platform_counts.values(), color=['red', 'blue', 'green', 'gold'])
ax[0].set_title('Количество фильмов на каждой платформе')
ax[0].set_ylabel('Количество фильмов')
ax[0].set_xlabel('Платформа')
ax[0].tick_params(axis='x', rotation=45)
ax[0].grid(axis='y', alpha=0.3) 

# Второй график - возрастное распределение
age_counts = df['Age'].value_counts().sort_index()  # Добавлена сортировка по индексу


#plt.figure(figsize=(12, 6))
#age_counts.plot(kind='bar', color=['red', 'blue', 'green', 'gold', 'purple', 'silver'], edgecolor='black')
ax[1].bar(age_counts.index, age_counts.values, color=['red', 'blue', 'green', 'gold', 'purple', 'silver'], edgecolor='black')
ax[1].set_title('Распределение по возрастам')  # plt.title('Распределение по возрастам')
ax[1].set_xlabel('Возраст')  #plt.xlabel('Возраст')
ax[1].set_ylabel('Частота') #plt.ylabel('Частота')
ax[1].tick_params(axis='x', rotation=45) #plt.xticks(rotation=45) #, ha="left"
ax[1].grid(axis='y', alpha=0.3) #plt.grid(axis='y', alpha=0.3)
plt.tight_layout() #plt.tight_layout()
plt.show() #plt.show()
```

```python
#dp["Year"] по декадам (десятилетиям)
#визуализация распределения фильмов по десятилетиям (bar)

# Создаем колонку с декадой
df['Decades'] = (df['Year'] // 10) * 10
 #числа на графике могут отображаться не все для этого преобразовываем в строку
df['Decade_label'] = df['Decade'].astype(str) + 's'

# Группируем по декадам и считаем количество
decades_counts = df['Decade_label'].value_counts().sort_index()

# Строим график
plt.figure(figsize=(12, 6))
decades_counts.plot(kind='bar', color='gold', edgecolor='black')
plt.title('Количество фильмов по декадам')
plt.xlabel('Декада')
plt.ylabel('Количество фильмов')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
