```python
def create_dashboard(plot_configs, vertical=True, dash_name="Дашборд"):
    """
    Создает дашборд с заданными конфигурациями графиков
    
    Parameters:
    plot_configs (list): Список конфигураций для каждого subplot
    vertical (bool): Ориентация дашборда (True - вертикально, False - горизонтально)
    dash_name (str): Название дашборда
    """
    # Определяем размерность сетки в зависимости от ориентации
    if vertical:
        nrows, ncols = len(plot_configs), 1
        figsize = (15, 5 * len(plot_configs))
    else:
        nrows, ncols = 1, len(plot_configs)
        figsize = (8 * len(plot_configs), 10)
    
    # Создаем фигуру и оси
    fig, axes = plt.subplots(nrows, ncols, figsize=figsize)
    fig.suptitle(dash_name, fontsize=16)
    
    # Если один график, преобразуем в массив для единообразия
    if len(plot_configs) == 1:
        axes = [axes]
    elif not vertical:
        axes = axes.flatten()
    
    # Строим графики с общими настройками
    for i, (ax, config) in enumerate(zip(axes, plot_configs)):
        ax.bar(config['labels'], config['data'], color=config['colors'])
        ax.set_title(config['title'])
        ax.set_ylabel(config['ylabel'])
        ax.set_xlabel(config['xlabel'])
        ax.tick_params(axis='x', rotation=45)
        ax.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Подготовка данных
platform_data = [
    ('Netflix', 'red'), 
    ('Hulu', 'blue'), 
    ('Prime Video', 'green'), 
    ('Disney+', 'gold')
]

age_counts = df['Age'].value_counts().sort_index()

# Конфигурации графиков
plot_configs = [
    {
        'data': [df[platform].sum() for platform, color in platform_data],
        'labels': [platform for platform, color in platform_data],
        'colors': [color for platform, color in platform_data],
        'title': 'Количество фильмов на каждой платформе',
        'ylabel': 'Количество фильмов',
        'xlabel': 'Платформа'
    },
    {
        'data': age_counts.values,
        'labels': age_counts.index,
        'colors': ['red', 'blue', 'green', 'gold', 'purple', 'silver'],
        'title': 'Распределение по возрастам',
        'ylabel': 'Частота',
        'xlabel': 'Возраст'
    }
]

# Использование функции
# Вертикальная ориентация (по умолчанию)
#create_dashboard(plot_configs, vertical=True)

# Горизонтальная ориентация
create_dashboard(plot_configs, vertical=False)

# Можно добавить еще графики
extended_plot_configs = plot_configs + [
    {
        'data': [df[df['Year'] >= 2000].shape[0], df[df['Year'] < 2000].shape[0]],
        'labels': ['После 2000', 'До 2000'],
        'colors': ['blue', 'red'],
        'title': 'Распределение по годам',
        'ylabel': 'Количество фильмов',
        'xlabel': 'Период'
    }
]

create_dashboard(extended_plot_configs)
```

```python
from typing import List, Dict, Any

def create_dashboard(plot_configs: List[Dict[str, Any]], vertical: bool = True, dash_name: str = "Дашборд") -> None:
    """
    Создает дашборд с заданными конфигурациями графиков
    
    Parameters:
    plot_configs: Список словарей с конфигурациями для каждого subplot
    vertical: Ориентация дашборда
    dash_name: Название дашборда
    """
    # Проверка во время выполнения
    if not isinstance(plot_configs, list):
        raise TypeError("plot_configs должен быть списком")
    
    for config in plot_configs:
        if not isinstance(config, dict):  # dict вместо 'dict'
            raise TypeError("Каждый элемент plot_configs должен быть словарем")
    
    # Определяем размерность сетки в зависимости от ориентации
    if vertical:
        nrows, ncols = len(plot_configs), 1
        figsize = (15, 5 * len(plot_configs))
    else:
        nrows, ncols = 1, len(plot_configs)
        figsize = (8 * len(plot_configs), 10)
    
    # Создаем фигуру и оси
    fig, axes = plt.subplots(nrows, ncols, figsize=figsize)
    fig.suptitle(dash_name, fontsize=16)
    
    # Если один график, преобразуем в массив для единообразия
    if len(plot_configs) == 1:
        axes = [axes]
    elif not vertical:
        axes = axes.flatten()
    
    # Строим графики с общими настройками
    for i, (ax, config) in enumerate(zip(axes, plot_configs)):
        ax.bar(config['labels'], config['data'], color=config['colors'])
        ax.set_title(config['title'])
        ax.set_ylabel(config['ylabel'])
        ax.set_xlabel(config['xlabel'])
        ax.tick_params(axis='x', rotation=45)
        ax.grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Подготовка данных
platform_data = [
    ('Netflix', 'red'), 
    ('Hulu', 'blue'), 
    ('Prime Video', 'green'), 
    ('Disney+', 'gold')
]

age_counts = df['Age'].value_counts().sort_index()

# Конфигурации графиков
plot_configs = [
    {
        'data': [df[platform].sum() for platform, color in platform_data],
        'labels': [platform for platform, color in platform_data],
        'colors': [color for platform, color in platform_data],
        'title': 'Количество фильмов на каждой платформе',
        'ylabel': 'Количество фильмов',
        'xlabel': 'Платформа'
    },
    {
        'data': age_counts.values,
        'labels': age_counts.index,
        'colors': ['red', 'blue', 'green', 'gold', 'purple', 'silver'],
        'title': 'Распределение по возрастам',
        'ylabel': 'Частота',
        'xlabel': 'Возраст'
    }
]

# Использование функции
# Вертикальная ориентация (по умолчанию)
#create_dashboard(plot_configs, vertical=True)

# Горизонтальная ориентация
#create_dashboard(plot_configs, vertical=False)

# Можно добавить еще графики
extended_plot_configs = plot_configs + [
    {
        'data': [df[df['Year'] >= 2000].shape[0], df[df['Year'] < 2000].shape[0]],
        'labels': ['После 2000', 'До 2000'],
        'colors': ['blue', 'red'],
        'title': 'Распределение по годам',
        'ylabel': 'Количество фильмов',
        'xlabel': 'Период'
    }
]

create_dashboard(extended_plot_configs)
```

```python
#dp["Year"] по декадам (десятилетиям)
#визуализация распределения фильмов по десятилетиям (bar)

# Создаем колонку с декадой
df['Decades'] = (df['Year'] // 10) * 10
 #числа на графике могут отображаться не все для этого преобразовываем в строку
df['Decade_label'] = df['Decade'].astype(str) + 's'

# Группируем по декадам и считаем количество
decades_counts = df['Decade_label'].value_counts().sort_index()

# Строим график
plt.figure(figsize=(12, 6))
decades_counts.plot(kind='bar', color='gold', edgecolor='black')
plt.title('Количество фильмов по декадам')
plt.xlabel('Декада')
plt.ylabel('Количество фильмов')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
