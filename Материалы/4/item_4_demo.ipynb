```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Формирование данных с np
x = np.linspace(0, 10, 50)
y1 = x
y2 = x**2

# 2 подграфика (1 строкаб 2 столбца)
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Левый подграфик (y=x)
axes[0].plot(x, y1, color="blue", label="y=x")
axes[0].set_title("Линейная функция")
axes[0].set_xlabel("Ось Х")
axes[0].set_ylabel("Ось Y")
axes[0].legend()
axes[0].grid(True)

# Правый подграфик (y = x**2)
axes[1].plot(x, y1, color="red", linestyle="--", label="y=x**2")
axes[1].set_title("Квадратичная функция")
axes[1].set_xlabel("Ось Х")
axes[1].set_ylabel("Ось Y")
axes[1].legend()
axes[1].grid(True)

# Общий заголовок результульрующего графика (Output data)
fig.suptitle("2 подграфика на одном графике", fontsize=14)
fig.tight_layout()

# Сохраним в PNG и SVG
fig.savefig("subplots_example.png", dpi=300)
fig.savefig("subplots_example.svg")

plt.show()
print("Файлы сохранены subplots_example.png и subplots_example.svg")
```
