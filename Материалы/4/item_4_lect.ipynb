```python
# Подготовка данных
x = np.linspace(-100, 100, 11)   # 11 точек
y1 = x
y2 = x**2
y3 = np.sqrt((x**5).astype(complex))

# Построение линейных диаграмм с разными стилями
plt.plot(x, y1, color="blue", linestyle="-", marker="o", linewidth=2, label="y = x")
plt.plot(x, y2, color="red", linestyle="--", marker="s", linewidth=2, label="y = x^2")
#plt.plot(x, y3, color="green", linestyle=":", marker="^", linewidth=2, label="y = √x")

plt.plot(x, y3.real,color="green", linestyle="solid", marker="^", linewidth=2, label="real y = √x^5")
plt.plot(x, y3.imag,color="yellow", linestyle=":", marker="v", linewidth=2, label="imag y = √x^5")

# 3) Оформление
plt.title("Примеры линейных диаграмм")
plt.xlabel("ось X")
plt.ylabel("ось Y")
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Подготовка данных
#np.random.seed(0)             # фиксируем генерацию случайных чисел
n = 200 # количество точек
#x = np.random.rand(n) * 2000 - 1000   # 50 случайных точек по оси X (от 0 до 10)
x = np.linspace(-10, 10, n) + np.random.rand(n)
y = 1/(x+np.random.randn(n))  # линейная зависимость с шумом
sizes = np.random.randint(20, 200, n)  # случайные размеры точек
colors = np.random.rand(n)             # случайные цвета

# Построение точечной диаграммы
plt.scatter(x, y, s=sizes, c=colors, alpha=0.7, cmap="viridis")

# Оформление
plt.title("Точечная диаграмма")
plt.xlabel("ось X")
plt.ylabel("ось Y")
plt.colorbar(label="цветовая шкала")
plt.grid(True)
plt.show()
```

```python
# Категории и значения
categories = ["A", "B", "C", "D", "E"]
values = [5, 7, 3, 8, 6]

# Построение вертикальной столбчатой диаграммы
plt.bar(categories, values, color="skyblue", edgecolor="black")

# Добавление подписей
plt.title("Столбчатая диаграмма")
plt.xlabel("Категории")
plt.ylabel("Значения")
plt.grid(axis="y")  # сетка только по оси Y
plt.show()
```

```python
import matplotlib.pyplot as plt

# построим визуализацию линейного графика
x = [i for i in range(6)]
y1 =  [i for i in x] # y = x
y2 =  [i**2 for i in x] # y = x**2

# построим визуализацию и добавим легенду и аннотацию
# на аннотации добавим выделенную точку с координатами xy (2,4)

# Создаём Figure и Axes
fig, ax = plt.subplots(figsize=(6, 4))

# Строим несколько графиков
ax.plot(x, y1, label="y = x")
ax.plot(x, y2, label="y = x^2")

# Добавляем элементы оформления
ax.set_title("Основные элементы диаграммы\n")
ax.set_xlabel("ось X")
ax.set_ylabel("ось Y")
ax.grid(True)                   # сетка
ax.legend()                     # легенда
ax.annotate("выделенная точка", 
            xy=(2, 4),    # точка для выделения
            xytext=(2, 1.2),    # подпись рядом
            arrowprops=dict(arrowstyle="->"))

plt.show()
```

```python
# Секторная диаграмма с выделением сегмента
categories = ["A", "B", "C", "D", "E", "F", "G", "H", "I"]
values = [20, 10, 10, 10, 10, 15, 5, 10, 10]
explode = [0.1, 0, 0.3, 0, 1.5, 0, 0, 0, 0]  # смещение для первой категории
plt.pie(values, labels=categories, autopct="%.1f%%",
        startangle=90, explode=explode, shadow=True)
plt.title("С выделением сектора")
plt.show()
```

```python
# Построение вертикальной столбчатой диаграммы
plt.grid(axis="y")
plt.bar(categories, values, color="skyblue", edgecolor="black")
```

```python
# Построение горизонтальной столбчатой диаграммы
plt.grid(axis="x")
colors = ["red" if cat == "A" else "skyblue" for cat in categories]
plt.barh(categories, values, color=colors, edgecolor="black")
```
